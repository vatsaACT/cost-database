```python
from sqlalchemy import create_engine
import pandas as pd
import openpyxl
```

```python
engine = create_engine(
    "postgresql://postgres:mypassword@localhost:5432/postgres")

conn = engine.connect()
print("Connected to PostgreSQL")
conn.close()
```

```python
df = pd.read_excel("GaN Inverter BOM_Elec.xlsx")

df.to_sql(name="GaN_eBOM",
            con=engine,
            if_exists="replace",
            index=False)
```

```python
query = """
SELECT *
FROM "GaN_eBOM"
LIMIT 5;
"""

df_query = pd.read_sql(query, con=engine)
```
